Correlational Analysis

```python
import seaborn as sns
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize = (40,40))
g = sns.heatmap(df[top_corr_features].corr(), annot = True, cmap = 'RdYlGn')
```

```python
#seperating the data and outcome
Y = df['Diabetes_012']
X = df.drop(columns='Diabetes_012',axis=1)
X.shape
#Y.shape
```

```python
Y
```

```python
np.unique(Y, return_counts=True)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
Y=np.array(Y)
Y = to_categorical(Y)
Y
```

Split Train and Test

```python
Y.shape
```

```python
np.unique(Y)
```

```python
np.unique(Y, return_counts=True)
```

```python
Y.shape
X[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify=Y)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
y_train.shape
```

Training Model

```python
from tensorflow import keras
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.linear_model import Perceptron
from sklearn.tree import DecisionTreeClassifier
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU
from keras.optimizers import RMSprop
model = Sequential()
model.add(Dense(32, input_shape=(21,)))
model.add(LeakyReLU())
model.add(Dense(64))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(128))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(256))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(512))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(256))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(128))
model.add(LeakyReLU())
model.add(Dropout(0.3))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001) , metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
model.evaluate(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
predict = model.predict(x_test)
predict_class = np.argmax(predict, axis=1)
```

```python
np.unique(predict_class, return_counts=True)
```

```python
y_real = np.argmax(y_test, axis=1)
np.unique(y_real, return_counts=True)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_real, predict_class)
print("Accuracy:",accuracy_score(y_real, predict_class))
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(cm, cmap=plt.cm.Blues, alpha=0.3)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(x=j, y=i,s=cm[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```
